```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from torch import tensor
```

```python
# 这是你的邻接矩阵
adjacency_matrix = tensor([[1.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000],
                           [0.1574, 0.8426, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000],
                           [0.2088, 0.1646, 0.6266, 0.0000, 0.0000, 0.0000, 0.0000, 0.0000],
                           [0.5792, 0.1187, 0.1889, 0.1131, 0.0000, 0.0000, 0.0000, 0.0000],
                           [0.0294, 0.1052, 0.0469, 0.0276, 0.7909, 0.0000, 0.0000, 0.0000],
                           [0.0176, 0.2689, 0.0215, 0.0089, 0.6812, 0.0019, 0.0000, 0.0000],
                           [0.1691, 0.4066, 0.0438, 0.0416, 0.1048, 0.2012, 0.0329, 0.0000],
                           [0.0210, 0.0843, 0.0555, 0.2297, 0.0573, 0.0709, 0.2423, 0.2391]])

# 创建图对象
G = nx.from_numpy_array(np.array(adjacency_matrix))
```

```python
G[1][1]
```

```python
# 获取边的权重
edge_weights = nx.get_edge_attributes(G, 'weight')
# 过滤权重不为0的边
filtered_edges = [(u, v) for u, v, weight in G.edges(data='weight') if weight != 0]
```

```python
# 绘制图
pos = nx.spring_layout(G)
labels = {i: i + 1 for i in range(G.number_of_nodes())}

# 绘制节点
nx.draw(G, pos, with_labels=True, labels=labels, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', font_size=8, edge_color='gray')

# 绘制带权重的边
edge_widths = [edge_weights[edge]*2 for edge in filtered_edges]
nx.draw_networkx_edges(G, pos, edgelist=filtered_edges, width=edge_widths, edge_color=edge_widths, edge_cmap=plt.cm.Reds)

# 显示图形
plt.show()
```
